# TensorFlow Intro

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
observations = 5000

x = np.random.uniform(low = -10, high = 10, size = (observations, 1))
z = np.random.uniform(low = -10, high = 10, size = (observations, 1))
noise = np.random.normal(0,1,(observations,1))

input = np.column_stack((x, z))

target = 2*x - 3*z + 5 + noise

np.savez("TF_Intro", inputs = input, targets = target)
```

```python
training_data = np.load('TF_Intro.npz')
```

```python
input_size = 2
output_size = 1

model = tf.keras.Sequential([
    tf.keras.layers.Dense(output_size)
])

model.compile(optimizer = 'sgd', loss = 'mean_squared_error')

model.fit(training_data['inputs'], training_data['targets'], epochs=100, verbose = 4)
```

```python
model.layers[0].get_weights()
```

```python
model.predict_on_batch(training_data['inputs']).round(1)
```

```python
training_data['targets'].round(1)
```

```python
plt.plot(np.squeeze(model.predict_on_batch(training_data['inputs'])), np.squeeze(training_data['targets']))
```
