```python
import numpy as np
import pandas as pd


from google.colab import drive
drive.mount('/content/drive')
```

```python
load = pd.read_csv('/content/drive/MyDrive/Absenteeism_data.csv')
df = load.copy()
df.head()
df.info()
```

```python
df.describe()
```

```python
df = df.drop(['ID'], axis=1)
df_no_age = df.drop(['Age'], axis=1)

df_no_age.head()
df.head()
```

```python
unique_reasons = df['Reason for Absence'].unique()
unique_reasons.sort()
print(len(unique_reasons))
print(unique_reasons)
```

```python
reason_column = pd.get_dummies(df['Reason for Absence'], drop_first = True)
reason_column
```

```python
reason_column['check'] = reason_column.sum(axis=1)
reason_column
```

```python
print(reason_column['check'].sum(axis=0))
print(reason_column['check'].unique())
```

```python
df = df.drop(['Reason for Absence'], axis=1)
df.head()
```

```python
reason_type_1 = reason_column.loc[:, 1:14].max(axis=1)
reason_type_2 = reason_column.loc[:, 15:17].max(axis=1)
reason_type_3 = reason_column.loc[:, 18:21].max(axis=1)
reason_type_4 = reason_column.loc[:, 22:].max(axis=1)
```

```python
df = pd.concat([df, reason_type_1, reason_type_2, reason_type_3, reason_type_4], axis = 1)
```

```python
df.head()
```

```python
df.columns.values
```

```python
column_names = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason 1', 'Reason 2', 'Reason 3', 'Reason 4']
df.columns = column_names
```

```python
column_names_reordered = ['Reason 1', 'Reason 2', 'Reason 3', 'Reason 4', 'Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
df[['Reason 1', 'Reason 2', 'Reason 3', 'Reason 4']] = df[['Reason 1', 'Reason 2', 'Reason 3', 'Reason 4']].astype(int)
df = df[column_names_reordered]
df.head()
```

```python
df_checkpoint_1 = df.copy() # Checkpoint 1 has the reason categories preprocessed
```

```python
type(df_checkpoint_1['Date'][0])
```

```python
df_checkpoint_1['Date'] = pd.to_datetime(df_checkpoint_1['Date'], format = '%d/%m/%Y')
type(df_checkpoint_1['Date'][0])
```

```python
months_list = []
```

```python
for i in range(df_checkpoint_1.shape[0]):
  months_list.append(df_checkpoint_1['Date'][i].month)
```

```python
df_checkpoint_1['Months'] = months_list
df_checkpoint_1.head()
```

```python
def date_to_weekday(date_value):
  return date_value.weekday()
df_checkpoint_1['Day of the Week'] = df_checkpoint_1['Date'].apply(date_to_weekday)
df_checkpoint_1.head()
```

```python
df_check2 = df_checkpoint_1.copy()
```

```python
print(df_check2['Education'].value_counts())

df_check2["Education"] = df_check2["Education"].map({1:0, 2:1, 3:1, 4:1})

print(df_check2['Education'].value_counts())
```
