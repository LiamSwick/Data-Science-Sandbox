```r
data = rnorm(100,mean=0,sd=1)
A = matrix(data=data,nrow=10,ncol=10)#Matrix containing the random (normal) data. In this case, of dim 10,10
k = 3
data2 = 1/(k^0.5)*rnorm(50,0,1)
B = matrix(data=data2,nrow=10,ncol=5)
proj_mat = A%*%B
for (i in length(B)){
    for (j in length(B)){
        if (i==j){
            
        epsilon = abs(A[i]/(proj_mat[j]-1))
        }
    }
}
epsilon
proj_mat
```
