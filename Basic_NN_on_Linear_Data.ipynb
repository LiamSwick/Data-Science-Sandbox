```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
observations = 500000
x = np.random.uniform(low=-10, high=10, size=(observations,1))
z = np.random.uniform(-10,10, size = (observations, 1))

inputs = np.column_stack((x,z))
print(inputs.shape)
```

```python
noise = np.random.normal(0,1,size = (observations,1))

target = (3*x) + (2*z) + 5 + noise

print(target.shape)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x, z, target)

ax.set_xlabel('x')
ax.set_ylabel('z')
ax.set_zlabel('target')

ax.view_init(elev=10, azim=250)

plt.show()
```

```python
init_range = 0.1

weights = np.random.uniform(-init_range, init_range, size = (2,1))

biases = np.random.uniform(-init_range, init_range, size = 1)

print(weights)
print(biases)
```

```python
learn_rate = 0.05
```

```python
for i in range(120):
  outputs = np.dot(inputs,weights) + biases
  delta = outputs - target
  loss = np.sum(np.sum(delta**2)) /(2*observations)
  print(loss)

  deltas_scaled = delta / observations
  weights = weights - (learn_rate * np.dot(inputs.T,deltas_scaled))
  biases = biases - (learn_rate * np.sum(deltas_scaled))
```

```python
print(weights, biases)
```

```python
plt.plot(outputs, target)
plt.xlabel('outputs')
plt.ylabel('target')
plt.show()
```
